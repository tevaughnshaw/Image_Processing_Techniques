# Cat Noisy Image Test

In this notebook, I will be testing opening a sample folder of cat images, adding random noise to them and saving the noisy images to a new folder.

```python
import cv2
import os
import numpy as np
```

```python
input_folder = 'cat-test'
output_folder = 'cat-noisy'
```

```python
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
# Iterate through images and add noise
for image in os.listdir(input_folder):
    if image.endswith('.jpg'):
        image_path = os.path.join(input_folder, image)
        original_image = cv2.imread(image_path)

        # Gaussian aka White noise
        mean = 0
        variance = 100
        sigma = np.sqrt(variance)
        noise = np.random.normal(mean, sigma, original_image.shape).astype(np.uint8)
        #  Apply noise to image
        noisy_image = cv2.add(original_image, noise)

        # Save noisy image to output folder
        noisy_image_path = os.path.join(output_folder, image)
        cv2.imwrite(noisy_image_path, noisy_image)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, layout='constrained', sharey=True)

axes[0].imshow(original_image[::-1], origin = 'lower')
axes[0].set_title('Original Image', fontsize = 15)
axes[1].imshow(noisy_image[::-1], origin = 'lower')
axes[1].set_title('Noisy Image',fontsize=15)

plt.suptitle('Comparison of Original and Noisy Image', y=.85,fontsize=20)
plt.savefig('noisy-cat-example.png')
plt.show()
```
